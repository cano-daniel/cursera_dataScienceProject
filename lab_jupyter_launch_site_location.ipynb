```python
## Task 1: Mark all launch sites on a map
import folium
import webbrowser
import tempfile
import os

def show_map(map_object):
    """
    Saves a Folium map to a temporary HTML file and opens it
    automatically in your default web browser.
    """
    # Create a temporary file
    tmp = tempfile.NamedTemporaryFile(suffix=".html", delete=False)
    
    # Save the map to it
    map_object.save(tmp.name)
    
    # Open it in the default browser
    webbrowser.open('file://' + os.path.realpath(tmp.name))


# ---- Example usage ----
m = folium.Map(location=[6.24, -75.57], zoom_start=12)
show_map(m)

```

```python
# Task 2: Mark the success/failed launches for each site on the map
spacex_df.head()
#new datafrem wich has the average of thelcolumn class and label is as sucches rate

success_rate_df = spacex_df.groupby('Launch Site').agg(
    success_rate=('class', 'mean'),
    Lat=('Lat', 'first'),  # Lat and Long are constant for each site, so 'first' or 'mean' works
    Long=('Long', 'first')
).reset_index()

#success_rate_df
#color maping tranformation 
def color_map(x:float) -> str:
    """
    input: the success rate
    output: color in format html

    does a continous mapping of the colors grean and red in the rgb space and the returns #rrggbb
    """
    return '#{:02x}{:02x}{:02x}'.format(int((1-x)*255), int(x*255), 0)

#task 3 iterate and givew a color with te sucches rate inn mind
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
circle_radius = 150 
circle_angle =   (circle_radius/6378137)*(180/3.1415)
# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label
for index, row in success_rate_df.iterrows():
    circle = folium.Circle(
        [row['Lat'], row['Long']],
        radius=circle_radius,
        popup=f"Success Rate: {row['success_rate']:.2f}",
        color= color_map(row['success_rate']),
        fill=True,
        fill_color= color_map(row['success_rate']),
        fill_opacity=0.4
    )
    marker = folium.Marker(
    [row['Lat']-circle_angle, row['Long']+circle_angle],
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color:{color_map(row["success_rate"])};"><b>{row["Launch Site"]}</b></div>'
        )
    )
    circle.add_child(folium.Popup(row['Launch Site']))
    site_map.add_child(circle)
    site_map.add_child(marker)

#add a line on the equator
equator = [[-0.0, -180.0], [-0.0, 180.0]]
site_map.add_child(folium.PolyLine(locations=equator, color='cyan'))
# Show the map
show_map(site_map)
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
show_map(site_map)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
coastline_lat, coastline_lon = 28.5618, -80.56765
launch_site_lat, launch_site_lon = success_rate_df.loc[0,'Lat'], success_rate_df.loc[0,'Long']
print(launch_site_lat, launch_site_lon) 
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_coastline = calculate_distance(coastline_lat, coastline_lon, launch_site_lat, launch_site_lon)
print(distance_coastline)

```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
city_lat, city_lon = 28.07856, -80.60806
railway_lat, railway_lon = 28.57175, -80.58524
highway_lat, highway_lon = 28.56666, -80.5713
# Draw a line between the marker to the launch site
for loc in (city_lat, city_lon), (railway_lat, railway_lon), (highway_lat, highway_lon):
    min_distance = float('inf')
    mindistance_index = 0
    for index, row in success_rate_df.iterrows():
        distance = calculate_distance(row['Lat'], row['Long'], loc[0], loc[1])
        if distance < min_distance:
            min_distance = distance
            mindistance_index = index
    
    line = folium.PolyLine(
        locations=[(success_rate_df.loc[mindistance_index, 'Lat']  , success_rate_df.loc[mindistance_index, 'Long']), (loc[0], loc[1])], 
        color='cyan', 
        weight=2,
        stroke=True,
        opacity=0.5
        )

    # lines=folium.PolyLine(locations=coordinates, weight=1)
    site_map.add_child(line)

show_map(site_map)
    
```
